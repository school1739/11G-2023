```python
import numpy as np
import pandas as pd
from datetime import datetime, date
import matplotlib.pyplot as plt
```

```python
random_seed=5432
```

```python
# генератор случайного временного ряда
np.random.seed(random_seed)
the_series1=pd.Series(np.random.randn(30),
                      index=pd.date_range("2023-01-01", "2023-01-30"))
the_series1
```

```python
# отображение кумулятивной суммы
cum_sum=the_series1.cumsum()
cum_sum.plot()
```

```python
cum_sum2=pd.DataFrame(cum_sum)
cum_sum2.plot()
```

```python
the_Series2=pd.DataFrame(np.random.randn(30, 2),
                         index=cum_sum2.index, columns=["Series 1", "Series 2"])
the_Series2.head()
```

```python
the_Series2.plot()
```

```python
the_massive=np.arange(0., 3., 2.0)
labels=["Сплошная", "Пунктир", "Точки", "Точка-пунктир"]
styles=pd.DataFrame({0: the_massive,
                     1: the_massive**1.5,
                     2: the_massive**2,
                     3: the_massive**3})
x_plot=styles.plot(style=["r-", "b--", "g.", "y.-"])
```
